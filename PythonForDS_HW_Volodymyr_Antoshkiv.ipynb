```python
from numpy import mean, median, std, percentile, fft, abs, argmax

def Mean(data):
    return data.mean()

def Median(data):
    return data.median()

def Std(data):
    return data.std()

def IQR(data):
    return percentile(data, 75) - percentile(data, 25)

def Min(data):
    return data.min()

def Max(data):
    return data.max()

def Length(data):
    return len(data)

def DominantFrequency(data):
    w = fft.fft(data)
    freqs = fft.fftfreq(len(data))
    i = argmax(abs(w))
    dom_freq = freqs[i]
    dom_freq_hz = abs(dom_freq * 32.0)
    return dom_freq_hz
```
